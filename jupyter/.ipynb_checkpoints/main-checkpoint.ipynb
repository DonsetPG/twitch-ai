```python
import twitch
import pandas as pd
import seaborn as sns
from pytimeparse.timeparse import timeparse
```

```python
with open("config.txt") as config:
    oauth,clientid=config.read().split(',')
```

```python
helix = twitch.Helix(clientid)
```

```python
def get_videos_from_user(user_name, limit=float('inf')):
    c=0
    videos=[]
    for video in helix.user(user_name).videos():
        if(c<limit):
            videos.append(video)
            c+=1
        else:
            break
    return videos
```

```python
def get_comments_from_video(video, limit=float('inf')):
    c=0
    comments=[]
    for comment in video.comments:
        if(c<limit):
            comments.append(comment)
            c+=1
        else:
            break
    return comments
    
```

```python
comments=get_comments_from_video(helix.video('554099089'), limit=10000)
```

```python
df=pd.DataFrame.from_dict({'content_offset_seconds':[comment.data['content_offset_seconds'] for comment in comments]})
sns.distplot( df["content_offset_seconds"], bins=100 )
```

```python
videos=get_videos_from_user("sardoche")
```

```python
df=pd.DataFrame.from_dict({'view_count'
                            :[video.data['view_count'] for video in videos],
                           'duration_seconds'
                            :[timeparse(video.data['duration']) for video in videos]
                          })
sns.regplot(x=df["view_count"], y=df["duration_seconds"])
```

```python
streamers=list(pd.read_csv('twitch_streamers.csv')['Channel'])
```

```python
helix.users()
```

```python
all_videos=[]
for streamer in streamers:
    print("Downloading videos for", streamer)
    videos=get_videos_from_user(streamer, limit=100)
    all_videos+=videos
```

```python
len(all_videos)
```

```python
df=pd.DataFrame.from_dict({'view_count'
                            :[video.data['view_count'] for video in all_videos],
                           'duration_seconds'
                            :[timeparse(video.data['duration']) for video in all_videos]
                          })
sns.regplot(x=df["view_count"], y=df["duration_seconds"])
```
